# **Q3 : Analyse d’Apprentissage Non Supervisé (Unsupervised Learning).**

```r
#remplir les variable num qui centient des na avec leur median
wealth_imputed <- wealth

num_cols <- sapply(wealth_imputed, is.numeric)

for (col in names(wealth_imputed)[num_cols]) {
  wealth_imputed[[col]][is.na(wealth_imputed[[col]])] <-
    median(wealth_imputed[[col]], na.rm = TRUE)
}
```

```r
data_unsup <- wealth_imputed[, sapply(wealth_imputed, is.numeric)]
data_unsup <- data_unsup[, !colnames(data_unsup) %in% "Target"]
```

```r
#normalisation
data_scaled <- scale(data_unsup)
```

# PCA (Analyse en composantes principales)

```r
pca_res <- prcomp(data_scaled, center = TRUE, scale. = TRUE)
summary(pca_res)
```

```r
plot(pca_res, type="l", main="Scree Plot PCA")
```

```r
install.packages("factoextra")
library(factoextra)
fviz_pca_var(pca_res,
             col.var = "contrib", # Couleur selon la contribution à l'axe
             gradient.cols = c("#00AFBB", "#E7B800", "#FC4E07"),
             repel = TRUE,        # Évite le chevauchement du texte
             title = "Cercle des corrélations - Variables (ACP)")
```

```r
pca_df <- as.data.frame(pca_res$x)
```

# Clustering (k-means)

```r
wss <- numeric()

for (k in 1:10) {
  set.seed(123)
  km <- kmeans(pca_df, centers = k, nstart = 25)
  wss[k] <- km$tot.withinss
}

plot(1:10, wss, type="b", pch=19,
     xlab="Nombre de clusters (k)",
     ylab="Somme des carrés intra-clusters (WSS)",
     main="Méthode du coude (Elbow)")
```

```r
library(cluster)

sil_width <- numeric()

for (k in 2:10) {
  set.seed(123)
  km <- kmeans(data_scaled, centers = k, nstart = 25)
  ss <- silhouette(km$cluster, dist(data_scaled))
  sil_width[k] <- mean(ss[, 3])
}

plot(2:10, sil_width[2:10], type="b", pch=19,
     xlab="Nombre de clusters (k)",
     ylab="Largeur moyenne de silhouette",
     main="Méthode de la silhouette")
```

> donc k=3

```r
kmeans_res <- kmeans(data_scaled, centers = 3, nstart = 25)
```

```r
pca_df$cluster <- kmeans_res$cluster

plot(pca_df$PC1, pca_df$PC2,
     col = pca_df$cluster,
     pch = 19,
     xlab = "PC1",
     ylab = "PC2",
     main = "Visualisation des clusters sur PCA")
```

# **Q6 — Supervised Learning (Target continue)**

```r
train_data$country <- as.factor(train_data$country)
test_data$country  <- as.factor(test_data$country)

train_data$urban_or_rural <- as.factor(train_data$urban_or_rural)
test_data$urban_or_rural  <- as.factor(test_data$urban_or_rural)

train_data$cluster <- as.factor(train_data$cluster)
test_data$cluster  <- as.factor(test_data$cluster)
```

```r
rec <- recipe(Target ~ ., data = train_data) %>%
  step_impute_median(all_numeric_predictors()) %>%
  step_normalize(all_numeric_predictors()) %>%
  step_dummy(all_nominal_predictors())
```

```r
rec_prep <- prep(rec, training = train_data)
```

```r
train_processed <- bake(rec_prep, new_data = train_data)
test_processed  <- bake(rec_prep, new_data = test_data)
```

```r
model_lm <- lm(Target ~ ., data = train_processed)

pred_lm <- predict(model_lm, test_processed)

rmse_lm <- rmse_vec(test_processed$Target, pred_lm)
r2_lm <- rsq_vec(test_processed$Target, pred_lm)

rmse_lm
r2_lm
```

```r
library(parsnip)
rf_model_simple <- rand_forest(
  mode = "regression",
  trees = 300
) %>%
  set_engine("ranger")
```

```r
rf_workflow_simple <- workflow() %>%
  add_recipe(rec) %>%
  add_model(rf_model_simple)
rf_fit_simple <- fit(rf_workflow_simple, data = train_data)
```

```r
rf_predictions_simple <- predict(rf_fit_simple, test_data) %>%
  bind_cols(test_data)
rf_metrics_simple <- rf_predictions_simple %>%
  metrics(truth = Target, estimate = .pred)

rmse_rf <- rf_metrics_simple %>% filter(.metric == "rmse") %>% pull(.estimate)
r2_rf <- rf_metrics_simple %>% filter(.metric == "rsq") %>% pull(.estimate)

rf_metrics_simple
```

```r
knn_spec <- nearest_neighbor(neighbors = 5) %>%
  set_engine("kknn") %>%
  set_mode("regression")
```

```r
knn_workflow <- workflow() %>%
  add_recipe(rec) %>%
  add_model(knn_spec)
```

```r
install.packages("kknn")
library(kknn)
```

```r
knn_fit <- knn_workflow %>%
  fit(data = train_data)

knn_results <- predict(knn_fit, test_data) %>%
  bind_cols(test_data)

knn_metrics <- metrics(knn_results, truth = Target, estimate = .pred)
rmse_knn <- knn_metrics %>%
  filter(.metric == "rmse") %>%
  pull(.estimate)
r2_knn <- knn_metrics %>%
  filter(.metric == "rsq") %>%
  pull(.estimate)
```

# **Q9 —analyse supervisée de classification**

```r
#split th data
test <- dataset[dataset$country %in% c("Ghana","Kenya","Nigeria"), ]
train <- dataset[!dataset$country %in% c("Ghana","Kenya","Nigeria"), ]
```

```r
install.packages("themis")
library(themis)
rec_class <- recipe(target_class ~ ., data = train) %>%
  step_impute_median(all_numeric_predictors()) %>%
  step_impute_mode(all_nominal_predictors()) %>%
  step_dummy(all_nominal_predictors()) %>%
  step_normalize(all_numeric_predictors()) %>%
  step_smote(target_class) #car on a imbalnced data
```

```r
install.packages("ranger")
rf_class_model <- rand_forest(
  mode = "classification",
  trees = 300
) %>%
  set_engine("ranger")
```

```r
rf_class_workflow <- workflow() %>%
  add_recipe(rec_class) %>%
  add_model(rf_class_model)
```

```r
rf_class_fit <- fit(rf_class_workflow, data = train)
```

```r
rf_class_predictions_class <- predict(rf_class_fit, test, type = "class") %>%
  bind_cols(test)

rf_class_predictions_prob <- predict(rf_class_fit, test, type = "prob") %>%
  bind_cols(test)

rf_class_metrics <- metrics(rf_class_predictions_class, truth = target_class, estimate = .pred_class)

acc_rf <- rf_class_metrics %>% filter(.metric == "accuracy") %>% pull(.estimate)
auc_rf <- roc_auc(rf_class_predictions_prob, truth = target_class, .pred_rich, event_level = "second") %>% pull(.estimate)

rf_class_metrics
```

```r
log_model <- logistic_reg(mode = "classification") %>%
  set_engine("glm")

log_workflow <- workflow() %>%
  add_recipe(rec_class) %>%
  add_model(log_model)

log_fit <- fit(log_workflow, data = train)

log_predictions_class <- predict(log_fit, test, type = "class") %>%
  bind_cols(test)

log_predictions_prob <- predict(log_fit, test, type = "prob") %>%
  bind_cols(test)

log_metrics <- metrics(log_predictions_class, truth = target_class, estimate = .pred_class)
acc_log <- log_metrics %>% filter(.metric == "accuracy") %>% pull(.estimate)
auc_log <- roc_auc(log_predictions_prob, truth = target_class, .pred_rich, event_level = "second") %>% pull(.estimate)

log_metrics
conf_mat(log_predictions_class, truth = target_class, estimate = .pred_class)
```

# **Q10 — Tuning des modèles de Machine Learning et stratégie d'ajustement**



```r
install.packages('xgboost')
library(xgboost)
```

```r
xgb_model <- boost_tree(
  mode = "classification",
  trees = tune(),
  tree_depth = tune(),
  learn_rate = tune(),
  loss_reduction = tune(),
  min_n = tune()
) %>%
  set_engine("xgboost")
```

```r
xgb_workflow <- workflow() %>%
  add_recipe(rec_class) %>%
  add_model(xgb_model)
```

```r
set.seed(123)
folds <- vfold_cv(train, v = 5)
```

```r
xgb_grid <- grid_regular(
  trees(range = c(200, 500)),
  tree_depth(range = c(3, 8)),
  learn_rate(range = c(0.01, 0.3)),
  loss_reduction(range = c(0.001, 0.1)),
  min_n(range = c(5, 20)),
  levels = 4
)
```

```r
set.seed(123)
xgb_tune <- tune_grid(
  xgb_workflow,
  resamples = folds,
  grid = xgb_grid,
  metrics = metric_set(accuracy, roc_auc)
)

xgb_tune
```

```r
best_params <- select_best(xgb_tune, metric = "roc_auc")
best_params
```

```r
final_xgb_workflow <- finalize_workflow(xgb_workflow, best_params)

final_xgb_fit <- fit(final_xgb_workflow, data = train)
```

```r
xgb_predictions <- predict(final_xgb_fit, test, type = "prob") %>%
  bind_cols(predict(final_xgb_fit, test, type = "class")) %>%
  bind_cols(test)
xgb_metrics_results <- metrics(xgb_predictions, truth = target_class, estimate = .pred_class)
acc_xgb <- xgb_metrics_results %>% filter(.metric == "accuracy") %>% pull(.estimate)
xgb_metrics_results
```

```r
auc_xgb <- roc_auc(xgb_predictions, truth = target_class, .pred_rich, event_level = "second") %>% pull(.estimate)
conf_mat(xgb_predictions, truth = target_class, estimate = .pred_class)
```
